```python
#2.Data Transformations :- cleaning, removing etc.,
#check for Null values
customer_data.isnull().sum()
```

```python
#checking for unique
customer_data.duplicated().sum()
```

```python
#Filling empty emails by their names
customer_data['email'] = customer_data['email'].fillna(customer_data['name'].str.replace(" ",'') + '@example.com')
customer_data
```

```python
customer_data.loc[35]
```

```python
#changing case into lower
customer_data['email'] = customer_data['email'].str.lower()
```

```python
customer_data.loc[35]
```

```python
customer_data['name'] = customer_data['name'].str.title()
```

```python
#handling phone number
customer_data['phone'] = customer_data['phone'].astype(str)
customer_data['phone'] = customer_data['phone'].str.replace('-','')
```

```python
#Load into csv 
customer_data.to_csv('cleaned_customer_data.csv', index = False)
```
