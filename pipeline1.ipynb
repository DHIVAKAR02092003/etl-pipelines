```python
def format_phone(df):
    length = len(df['phone'])
    for i in range(length):
        if len(df['phone'][i]) > 10 or len(df['phone'][i]) < 10:
            df['phone'][i] = 0 
        elif df['code'][i] == 'IN':
            df['phone'][i] = '+91 ' + str(df['phone'][i])
        elif df['code'][i] == 'AE':
            df['phone'][i] = '+971 ' + str(df['phone'][i])
        elif df['code'][i] == 'FM':
            df['phone'][i] = '+691 ' + str(df['phone'][i])
        elif df['code'][i] == 'PW':
            df['phone'][i] = '+680 ' + str(df['phone'][i])
        else: 
            df['phone'][i] = '+1 ' + str(df['phone'][i])
    return df

format_phone(df)
```
