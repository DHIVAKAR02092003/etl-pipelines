```python
#Loading a Data
import pandas as pd 
customer_data = pd.read_csv(r"C:\Users\Dhivakar\pipelines\us_customer_data 1.csv")
customer_data
```

```python
#transform
customer_data['name'] = customer_data['name'].str.strip().str.title()
customer_data['email'] = customer_data['email'].str.strip().str.lower()
#digits only
customer_data['phone'] = customer_data['phone'].astype(str).str.replace(r'\D', '', regex = True)
```

```python
#extracting codes
customer_data['country_code'] = customer_data['address'].str.slice(-8,-6)
customer_data.head()
```

```python
customer_data[customer_data['customer_id'] == 400]
```

```python
def format_phone(row):
    phone = str(row['phone']).strip()
    
    phone_digits = ''.join(filter(str.isdigit, phone))

    if len(phone_digits) != 10:
        return 'Invalid'

    code = row['country_code']
    if code == 'IN':
        return '+91 ' + phone_digits
    elif code == 'AE':
        return '+971 ' + phone_digits
    elif code == 'FM':
        return '+691 ' + phone_digits
    elif code == 'PW':
        return '+680 ' + phone_digits
    else:
        return '+1 ' + phone_digits


customer_data['phone'] = customer_data.apply(format_phone, axis=1)
```
