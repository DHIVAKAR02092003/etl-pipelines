```python
import pandas as pd 
df = pd.read_csv(r"C:\Users\Dhivakar\pipelines\transaction_data.csv")
df
```

```python
df.isnull().sum()
```

```python
#cleaning the data 
df['amount'] = pd.to_numeric(df['amount'], errors = 'coerce')

df = df.dropna(subset=['amount'])
df = df[df['amount'] > 0]
```

```python
high_value = df[df['amount'] > 1000]

mean_value = df['amount'].mean()
high_value_mean = df[df['amount'] > mean_value]
```

```python
high_value_combined = pd.concat([high_value, high_value_mean]).drop_duplicates()
```

```python
df['vip status'] = df['amount'].apply(lambda x: 'VIP' if x > 100 else 'Regular')
df
```
